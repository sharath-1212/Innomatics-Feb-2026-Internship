### Problem 1: Unique Words in a Sentence

```python
# Given sentence
sentence = "python is easy and python is powerful"

# Split sentence into words
words = sentence.split()

# Convert list to set to remove duplicates
unique_words = set(words)

# Display results
print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)
```

### Problem 2: Highest Salary from Employee Data

```python
# Dictionary with employee names and salaries
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

# Find employee with highest salary
highest_employee = max(employees, key=employees.get)

# Get the salary of that employee
highest_salary = employees[highest_employee]

print(f"Highest Salary: {highest_employee} - {highest_salary}")
```

### Problem 3: Find Maximum and Minimum Values

```python
numbers = [45, 22, 89, 10, 66]

# Assume first element as max and min
max_val = numbers[0]
min_val = numbers[0]

# Loop through the list
for num in numbers:
    if num > max_val:
        max_val = num
    if num < min_val:
        min_val = num

print("List:", numbers)
print("Max:", max_val)
print("Min:", min_val)
```

### Problem 4: Count Products Above 1000

```python
prices = [450, 1200, 899, 1500, 300]

count = 0

# Check each price
for price in prices:
    if price > 1000:
        count += 1

print("Products above 1000:", count)
```

### Problem 5: Calculate Attendance Percentage

```python
attendance = ["P", "P", "A", "P", "P"]

# Count number of Present days
present_count = attendance.count("P")

# Total days
total_days = len(attendance)

# Calculate percentage
percentage = (present_count / total_days) * 100

print("Attendance Percentage:", percentage)
```

### Problem 6: Remove Duplicate Phone Numbers

```python
phone_numbers = [9876543210, 9123456789, 9876543210]

# Convert list to set to remove duplicates
unique_numbers = set(phone_numbers)

print("Unique phone numbers:", unique_numbers)
```

### Problem 7: Count Character Frequency

```python
text = "python"

# Empty dictionary to store frequency
char_count = {}

# Count each character
for char in text:
    if char in char_count:
        char_count[char] += 1
    else:
        char_count[char] = 1

print(char_count)
```

### Problem 9: Check if a Key Exists in Dictionary

```python
employees = {
    "Ravi": 75000,
    "Anita": 68000
}

key = "Ravi"

# Check if key exists
if key in employees:
    print("Employee exists")
else:
    print("Employee does not exist")
```

### Problem 10: Calculate Average Marks

```python
marks = [70, 75, 80, 65]

# Calculate average
average = sum(marks) / len(marks)

print("Average Marks:", average)
```
