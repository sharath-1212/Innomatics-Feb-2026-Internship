## Problem 1: Social Media – Post Engagement Analyzer

```python

def post_engagement(likes_list):
    # Calculate total likes using loop
    total_likes = 0
    for like in likes_list:
        total_likes += like

    # Check status
    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"

    print("Total Likes:", total_likes)
    print("Post Status:", status)

# Example
post_engagement([200, 300, 250, 300])
post_engagement([195, 325, 145, 245])
```

## Problem 3: Agriculture – Rainfall Adequacy Checker

```python

def rainfall_check(rainfall_data, required_level):
    # Calculate average rainfall
    total = 0
    for rain in rainfall_data:
        total += rain
    average = total / len(rainfall_data)

    # Check adequacy
    if average >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"

    print("Average Rainfall:", average)
    print("Rainfall Status:", status)

# Example
rainfall_check([60, 80, 70, 78], 70)
```

## Problem 4: Social Media – Duplicate Account Detection

```python

def check_duplicates(usernames):
    # Using set to detect duplicates
    if len(usernames) != len(set(usernames)):
        result = "Yes"
    else:
        result = "No"

    print("Duplicate Accounts Found:", result)

# Example
check_duplicates(["user1", "user2", "user3", "user1"])
```

## Problem 6: Agriculture – Premium Crop Price Filter

```python

def premium_crops(prices):
    # Filter premium crops
    premium = []
    for price in prices:
        if price > 2000:
            premium.append(price)

    print("Premium Crops:", premium)

# Example
premium_crops([1500, 2500, 1800, 3200])
```

## Problem 7: System Monitoring – Application Health Checker

```python

def system_health(errors):
    # Determine system status
    if errors == 0:
        status = "Healthy"
    elif errors <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"

    print("Error Count:", errors)
    print("System Status:", status)

# Example
system_health(7)
```

## Problem 9: E-Learning – Student Attendance Eligibility System

```python

def attendance_eligibility(attendance_list):
    # Calculate attendance percentage
    present_days = 0
    for day in attendance_list:
        if day == 1:
            present_days += 1

    percentage = (present_days / len(attendance_list)) * 100

    # Check eligibility
    if percentage >= 75:
        status = "Eligible"
    else:
        status = "Not Eligible"

    print("Attendance Percentage:", percentage)
    print("Exam Eligibility:", status)

# Example (1 = Present, 0 = Absent)
attendance_eligibility([1,1,1,0,1,1,1,1,0,1])
```

## Problem 10: Smart Electricity Bill Analyzer

```python

def electricity_bill(units):
    # Calculate bill based on slabs
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + (units - 100) * 5
    else:
        bill = (100 * 3) + (100 * 5) + (units - 200) * 7

    # Usage classification
    if bill < 500:
        status = "Low Usage"
    elif bill <= 1500:
        status = "Moderate Usage"
    else:
        status = "High Usage"

    print("Units Consumed:", units)
    print("Total Bill:", bill)
    print("Usage Level:", status)

# Example
electricity_bill(250)
```
