```python
# Problem 1: Mobile Recharge Validation System

def validate_recharge():
    valid_plans = [199, 299, 399, 599]

    while True:
        amount = int(input("Enter recharge amount: ₹"))

        if amount < 50:
            print(" Recharge amount must be at least ₹50.")
        elif amount not in valid_plans:
            print(" Invalid plan. Available plans:", valid_plans)
        else:
            print(" Recharge successful!")
            break
validate_recharge()
```

```python
# Problem 2: Inventory Reorder Alert System

def check_inventory(stock_dict):
    for product, quantity in stock_dict.items():
        if quantity < 15:
            print(f" Reorder Alert for {product} (Stock: {quantity})")
        else:
            print(f" Stock OK for {product} (Stock: {quantity})")

inventory = {
    "Rice": 10,
    "Wheat": 25,
    "Oil": 12,
    "Sugar": 30
}

check_inventory(inventory)
```

```python
# Problem 3: Student Result Processing System

def process_result(marks):
    total = 0
    for mark in marks:
        total += mark

    average = total / len(marks)

    print("Average Marks:", average)

    if average >= 50:
        return "Pass"
    else:
        return "Fail"

student_marks = [60, 70, 55, 40, 65]
print("Result:", process_result(student_marks))
```

```python
# Problem 4: Cab Fare Estimator with Retry Option

def calculate_fare():
    base_fare = 50
    rate_per_km = 12

    while True:
        distance = float(input("Enter distance in km: "))
        peak = input("Is it peak hour? (yes/no): ").lower()

        fare = base_fare + (distance * rate_per_km)

        if peak == "yes":
            fare += fare * 0.25

        print(f"Estimated Fare: ₹{fare}")

        retry = input("Do you want to calculate again? (yes/no): ").lower()
        if retry != "yes":
            break
calculate_fare()
```

```python
# Problem 5: Employee Attendance Eligibility Checker

def check_eligibility(attendance_list):
    present_count = 0

    for day in attendance_list:
        if day == "P":
            present_count += 1

    percentage = (present_count / len(attendance_list)) * 100

    print("Attendance Percentage:", percentage)

    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"

attendance = ["P", "P", "A", "P", "P", "A", "P", "P"]
print("Eligibility:", check_eligibility(attendance))
```

```python
# Problem 6: Password Strength Checker

def check_password(password):
    special_chars = "@#$"
    has_digit = False
    has_special = False

    if len(password) < 8:
        return " Weak Password (Minimum 8 characters required)"

    for char in password:
        if char.isdigit():
            has_digit = True
        if char in special_chars:
            has_special = True

    if has_digit and has_special:
        return " Strong Password"
    else:
        return " Weak Password (Must contain at least one digit and one special character @#$)"

print(check_password("Test@123"))
print(check_password("hello@#"))
```
