```python

# 1. User Login Check
print("---- Task 1: User Login Check ----\n")
username = "admin"
password = "1234"

username_input = input("Enter username: ")
password_input = input("Enter password: ")

if username_input == username and password_input == password:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

```python
# 2. Pass / Fail Analyzer
print("\n---- Task 2: Pass / Fail Analyzer ----\n")

marks = [45, 78, 90, 33, 60]

pass_count = 0
fail_count = 0

for mark in marks:
    if mark >= 50:
        pass_count += 1
    else:
        fail_count += 1

print("Total Students:", len(marks))
print("Passed Students:", pass_count)
print("Failed Students:", fail_count)
```

```python
# 3. Simple Data Cleaner
print("\n---- Task 3: Simple Data Cleaner ----\n")

names = [" Alice ", "bob", " CHARLIE "]

cleaned_names = []

for name in names:
    cleaned_name = name.strip().lower()
    cleaned_names.append(cleaned_name)

print("Cleaned Names:", cleaned_names)
```

```python
# 4. Message Length Analyzer
print("\n---- Task 4: Message Length Analyzer ----\n")

messages = ["Hi", "Welcome to the platform", "OK"]

for msg in messages:
    length = len(msg)
    print(f"Message: '{msg}' | Length: {length} \n")
    
    if length > 10:
        print("-> Flag: Message longer than 10 characters")
```

```python
# 5. Error Message Detector
print("\n---- Task 5: Error Message Detector ----\n")

logs = ["INFO", "ERROR", "WARNING", "ERROR", "ERROR"]

error_count = 0

for log in logs:
    if log == "ERROR":
        error_count += 1

print("Total ERROR entries:", error_count)
```
