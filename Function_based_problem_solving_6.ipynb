## Problem 1: Smart Parking Lot Management System

```python
# Function to manage parking logs
def parking_system(capacity, logs):
    parked = 0
    peak = 0

    # Loop through vehicle logs
    for log in logs:
        if log == "IN":
            parked += 1
        elif log == "OUT" and parked > 0:
            parked -= 1

        # Track peak usage
        if parked > peak:
            peak = parked

    print("Currently Parked Vehicles:", parked)
    print("Peak Usage:", peak)

    # Capacity check
    if parked > capacity:
        print("Alert: Parking Exceeded Capacity!")
    else:
        print("Parking Status: Available")


# Sample Input
parking_system(50, ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"])
```

## Problem 2: Online Food Delivery Time Estimator

```python
# Function to estimate delivery time
def estimate_delivery(distance, traffic, weather):
    # Base time calculation (2 minutes per km)
    time = distance * 2

    # Traffic delay conditions
    if traffic == "High":
        time += 20
    elif traffic == "Medium":
        time += 10

    # Weather delay conditions
    if weather == "Rainy":
        time += 15
    elif weather == "Storm":
        time += 25

    print("Estimated Delivery Time:", time, "minutes")


# Sample Input
estimate_delivery(8, "High", "Rainy")
```

## Problem 3: Movie Theatre Seat Occupancy Analyzer

```python
# Function to analyze theatre occupancy
def theatre_occupancy(total_seats, booked_list):
    booked = len(booked_list)
    occupancy = (booked / total_seats) * 100

    print("Occupancy:", round(occupancy), "%")

    # Show status conditions
    if occupancy == 100:
        print("Show Status: Housefull")
    elif occupancy >= 75:
        print("Show Status: Almost Full")
    else:
        print("Show Status: Seats Available")

    # Suggest additional show if very high demand
    if occupancy >= 90:
        print("Suggestion: Open Additional Show")


# Sample Input
booked_seats = [1] * 150
theatre_occupancy(200, booked_seats)
```

## Problem 4: Cloud Server Load Classification System

```python
# Function to classify server load
def server_load(cpu_readings):
    total = 0

    # Calculate total CPU usage
    for value in cpu_readings:
        total += value

    average = total / len(cpu_readings)
    print("Average CPU Load:", round(average), "%")

    # Classification conditions
    if average < 50:
        print("Server Status: Normal")
    elif average <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")


# Sample Input
server_load([45, 60, 70, 85, 90])
```

## Problem 5: Smart Classroom Resource Usage Monitor

```python
# Function to monitor resource usage
def resource_monitor(usage_dict):
    overused = []

    # Check each resource
    for resource, hours in usage_dict.items():
        if hours > 8:  # Threshold for overuse
            overused.append(resource)

    # Output results
    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")


# Sample Input
usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

resource_monitor(usage)
```

## Problem 6: Online Event Registration Capacity Controller

```python
# Function to manage event registrations
def event_registration(capacity, registrations):
    if registrations <= capacity:
        print("Confirmed Registrations:", registrations)
        print("Waitlisted Users: 0")
        print("Registration Status: Open")
    else:
        waitlist = registrations - capacity
        print("Confirmed Registrations:", capacity)
        print("Waitlisted Users:", waitlist)
        print("Registration Status: Closed")


# Sample Input
event_registration(100, 105)
```
